```python
import pandas as pd
from selenium import webdriver # هذي المكتبة تفتح متصفح قوقل كروم
import requests # هذي المكتبة تجبلنا html في اي موقع
from bs4 import BeautifulSoup as bs
from selenium.webdriver.common.action_chains import ActionChains as act # هذي المكتبة راح تخلينا نقدر نضغط في الصفحة
```

```python
driver = webdriver.Chrome('C:\Program Files (x86)\chromedriver') # الملف اللي حملناه وتاكد انك حاطه في نفس المسار
driver.get('https://ajel.sa/local/') # 
try:
    for i in range(5):
        element= driver.find_element_by_xpath('//*[@id="moreCategorySubmenu"]').click()
except:
    print("An exception occurred")
    

x = driver.find_element_by_xpath('//*[@id="tab_category_1"]')
allUrls=list() 
for i in range(75): 
    x1= x.find_element_by_xpath('//*[@id="tab_category_1"]/div[{o}]/div[2]/div[1]/a'.format(o=i+1)) 
    allUrls.append(x1.get_attribute('href'))
driver.quit()
```

```python
def info(url):
    soup = bs(requests.get(url).content, 'html.parser')
    text = soup.find('div', attrs={"class":"p-block"}).get_text().replace("\n","").strip().replace("اقرأ أيضا:كورونا السعودية..","").replace(',','،')
    title = str(soup.find('div',attrs={"news-titles"}).find('h1')).replace('<h1>',"").replace('</h1>',"").replace(',','،')
    return[title,text]
```

```python
with open('ajil1.csv','w',encoding="utf-8") as file:
    file.write('Newsletter,Category,URl,Title,Text')
    file.write('\n')
    for url in allUrls:
        now = info(url)
        file.write('عاجل' + ',' + 'محلي' + ',' +url + ',' + now[0]+ ',' + now[1])
        file.write('\n')
```
